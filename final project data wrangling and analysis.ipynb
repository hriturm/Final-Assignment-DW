# 4. Filter the dataframe with only this columns ['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index'] and update the data frame.

```python
data_set = data_set[{'continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index'}]
data_set
```

# 5. Data Cleaning
## a. Remove all duplicates observations
## b. Find missing values in all columns
## c. Remove all observations where continent column value is missing
## Tip : using subset parameter in dropna
## d. Fill all missing values with 0

```python
#Remove all duplicates observations
data_set.drop_duplicates(keep = False, inplace = True)
data_set
```

```python
#Find missing values in all columns
data_set.isnull().sum()
```

```python
#Remove all observations where continent column value is missing
data_set.dropna(how = "all", subset = ["continent"], inplace = True)
data_set
```

```python
# Fill all missing values with 0
data_set = data_set.fillna(0)
data_set
```

# 6. Date time format :
## a. Convert date column in datetime format using pandas.to_datetime
## b. Create new column month after extracting month data from date column.

```python
#Convert date column in datetime format using pandas.to_datetime
data_set['date'] = pd.to_datetime(data_set['date'])
data_set
```

```python
#to check if the data type of date coloumn is changed
data_set.dtypes
```

```python
#Create new column month after extracting month data from date column.
data_set['Month'] = pd.DatetimeIndex(data_set['date']).month
data_set
```

# 7. Data Aggregation:
## a. Find max value in all columns using groupby function on'continent' column
## Tip: use reset_index() after applying groupby
## b. Store the result in a new dataframe named 'df_groupby'. (Use df_groupby dataframe for all further analysis)

```python
#Find max value in all columns using groupby function on'continent' column
# Store the result in a new dataframe named 'df_groupby'. (Use df_groupby dataframe for all further analysis)

df_groupby = data_set.groupby('continent').max().reset_index()
df_groupby.head()
```

# 8. Feature Engineering :
## a. Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to 'total_cases'

```python
df_groupby['total_deaths_to_total_cases'] = df_groupby['total_deaths'] / df_groupby['total_cases']
df_groupby.head()
```

# 9. Data Visualization :

## a. Perform Univariate analysis on 'gdp_per_capita' column by plotting histogram using seaborn dist plot.
## b. Plot a scatter plot of 'total_cases' & 'gdp_per_capita'
## c. Plot Pairplot on df_groupby dataset.
## d. Plot a bar plot of 'continent' column with 'total_cases' . Tip : using kind='bar' in seaborn catplot

```python
#Perform Univariate analysis on 'gdp_per_capita' column by plotting histogram using seaborn dist plot.
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.distplot(df_groupby['gdp_per_capita'])
```

```python
#Plot a scatter plot of 'total_cases' & 'gdp_per_capita'
x = df_groupby['total_cases']
y = df_groupby['gdp_per_capita']
plt.scatter(x, y, label = "Points", c = "Blue")

#set X and Y axis lable
plt.xlabel(" Total Cases")
plt.ylabel(" GDP Per Capital")
plt.title("Scatter Plot")
plt.show()
```

```python
#Plot Pairplot on df_groupby dataset.
sns.pairplot(df_groupby)
```

```python
# Plot a bar plot of 'continent' column with 'total_cases' . Tip : using kind='bar' in seaborn catplot¶
sns.catplot(x = "continent", y = "total_cases", data = df_groupby, kind = "bar" )
```
